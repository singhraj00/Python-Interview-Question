```python
def sort_arr(arr):
  for i in range(len(arr)):
    for j in range(len(arr)-i-1):
      if arr[j]>arr[j+1]:
        arr[j],arr[j+1]=arr[j+1],arr[j]
  return arr


sorted_arr = sort_arr([55,11,9,14])
sorted_arr
```

```python
def two_sum(arr,target):
  seen = {}
  for i,num in enumerate(arr):
    comp = target-num
    if comp in seen:
      return [seen[comp],i]
    seen[num]=i
  return []


two_sum([2,5,7,11],9)
```

```python
def valid_parenthesis(text):
  stack=[]
  mappings = {
      ']':'[',
      '}':'{',
      ')':'('
  }

  for char in text:
    if char in mappings:
      if not stack or stack.pop()!=mappings[char]:
        return False
    else:
      stack.append(char)
  return len(stack)==0


valid_parenthesis('({[]})')
```

```python
def long_substring(text):
  seen = {}
  max_len = 0
  left = 0
  for right,char in enumerate(text):
    if char in seen and seen[char]>=left:
      left = seen[char]+1
    else:
      max_len = max(max_len,right-left+1)
    seen[char]=right
  return max_len

long_substring('rahul rathour')
```

```python
def find_occurence(text):
  freq = {}
  for char in text:
    freq[char] = freq.get(char,0)+1
  return freq

find_occurence('rahul rathour')
```

```python

def my_decorator(func):
  def wrapper(*args,**kwargs):
    print('before function...')
    result = func(*args,**kwargs)
    print(result)
    print('after function....')
  return wrapper

@my_decorator
def my_function(a,b):
  return a+b

my_function(2,3)
```

```python
import time

def measure_time(func):
  def wrapper(*args,**kwargs):
    start = time.time()
    print('Function Executing Start...')
    result = func(*args,**kwargs)
    end = time.time()
    print('function execution end...')
    print(f'{func.__name__} takes time at {end-start:.3f} seconds')
  return wrapper

@measure_time
def my_function(a,b):
  time.sleep(5)
  return a+b

my_function(2,3)
```

```python
## in place swap
def rev_list(lst):
  left,right=0,len(lst)-1
  while left<right:
    lst[left],lst[right]=lst[right],lst[left]
    left+=1
    right-=1
  return lst

rev_list([1,2,3,4,5])
```

```python
## using recursion
def rev_list(arr):
  if len(arr)==0:
    return []
  return [arr[-1]] + rev_list(arr[:-1])

rev_list([1,2,3,4,5])
```

```python
def rev_list_inplace(arr):
  left,right=0,len(arr)-1
  while left<right:
    arr[left],arr[right]=arr[right],arr[left]
    left+=1
    right-=1
  return arr

rev_list_inplace([1,2,3,4,5])
```

```python
def rev_list_rec(arr):
  if len(arr)==0:
    return []
  return [arr[-1]] + rev_list_rec(arr[:-1])

rev_list_rec([1,2,3,4,5])
```

```python
def merge_two_sorted_list(a,b):
  i=j=0
  merged = []
  while i < len(a) and j<len(b):
    if a[i]<b[j]:
      merged.append(a[i])
      i+=1
    else:
      merged.append(b[j])
      j+=1
  merged += a[i:] + b[j:]
  return merged

merge_two_sorted_list([1,2,3],[4,5,6])
```

```python
def find_duplicates(arr):
  seen = set()
  duplicates = set()

  for num in arr:
    if num in seen:
      duplicates.add(num)
    else:
      seen.add(num)
  return list(duplicates)


find_duplicates([1,2,3,4,5,5,6,1,2])
```

```python
def highest_freq(lst):
  freq={}
  for x in lst:
    if x in freq:
      freq[x] += 1
    else:
      freq[x]=1

  max_count=0
  max_elem = None

  for key,value in freq.items():
    if value>max_count:
      max_count=value
      max_elem=key
  return max_elem,max_count

highest_freq([1,2,2,3,4,1,2,4,3])
```
