# 1. Longest Substring Without Repeating Characters

````python
def longestSubString(s):
  chars_set = set()
  max_len = 0
  left = 0

  for right in range(len(s)):
    while s[right] in chars_set:
      chars_set.remove(s[left])
      left+=1
    chars_set.add(s[right])
    max_len = max(max_len,right-left+1)
  return max_len


longestSubString('aabbbccd')
````

# 2. Valid Parentheses


````python
def validParentheses(s):
  stack=[]
  mappings={
      '}':'{',
      ']':'[',
      ')':'('
  }

  for ch in s:
    if ch in mappings:
      if not stack or stack.pop() != mappings[ch]:
        return False
    else:
      stack.append(ch)
  return len(stack)==0

validParentheses('{([])}')
````

# 3. Find the Index of the First Occurrence in a String
Given two strings `needle` and `haystack`, return the index of the first occurrence of needle in haystack, or `-1` if needle is not part of haystack.

## Example 1:
```
Input: haystack = "sadbutsad", needle = "sad"
Output: 0
Explanation: "sad" occurs at index 0 and 6.
The first occurrence is at index 0, so we return 0.
```
## Example 2:
```
Input: haystack = "leetcode", needle = "leeto"
Output: -1
Explanation: "leeto" did not occur in "leetcode", so we return -1.
```

````python
def findIndexFirstOcc(haystack,needle):
  return haystack.find(needle)

print(findIndexFirstOcc('sadbutsad','sad'))
print(findIndexFirstOcc('leetcode','leeto'))
````

````python
def strStr(haystack,needle):
  if not needle:
    return None
  n,m=len(haystack),len(needle)
  for  i in range(n-m+1):
    if haystack[i:i+m]==needle:
      return i
  return -1


strStr('sadbutsad','sad')
````

# 4. Valid Palindrome

````python
def validPalindrome(s):
  l,r = 0 , len(s)-1

  while l<r:
    while l<r and not s[l].isalnum():
      l+=1
    while l<r and not s[r].isalnum():
      r-=1

    if s[l].lower() != s[r].lower():
      return False

    l+=1
    r-=1

  return True

print(validPalindrome('A man, a plan, a canal: Panama'))
print(validPalindrome('race a car'))
print(validPalindrome(' '))
````

# 5. Reverse String

````python
def revString(lst):
  l,r = 0 ,len(lst)-1
  while l<r:
    lst[l],lst[r]=lst[r],lst[l]
    l+=1
    r-=1
  return lst

revString(['h','e','l','l','o'])
````

# 6. First Unique Character in a String

````python
def firstUniqueChar(s):
  freq = {}
  for ch in s:
    freq[ch]=freq.get(ch,0)+1
    if freq[ch]==1:
      print(ch)
      break

firstUniqueChar('bananac')
````

# 7. Greatest Common Divisor of Strings

For two strings `s` and `t`, we say `"t divides s"` if and only if `s = t + t + t + ... + t + t` (i.e., t is concatenated with itself one or more times).

Given two strings `str1` and `str2`, return the largest string x such that x divides both str1 and str2.

### Examples :

```
Input: str1 = "ABCABC", str2 = "ABC"
Output: "ABC"
Example 2:

Input: str1 = "ABABAB", str2 = "ABAB"
Output: "AB"
Example 3:

Input: str1 = "LEET", str2 = "CODE"
Output: ""
```

````python
def gcd(a,b):
  while b!=0:
    a,b=b,a%b
  return a

def gcdString(str1,str2):
  if str1 + str2 != str2 + str1:
    return ""
  gcd_len = gcd(len(str1),len(str2))
  return str1[:gcd_len]

print(gcdString('ABCABC','ABC'))
print(gcdString('ABABAB','AB'))
print(gcdString('LEET','CODE'))
````

# String Permutation

````python
from itertools import permutations
s='abc'

["".join(list(ch)) for ch in permutations(s)]
````

````python
def permutation(s):
  s = list(s)
  res = []

  def backtrack(i):
    if i==len(s):
      res.append("".join(s))
      return

    for j in range(i,len(s)):
      s[i],s[j]=s[j],s[i]
      backtrack(i+1)
      s[i],s[j]=s[j],s[i]

  backtrack(0)
  return res


permutation('abc')
````

# 9. String Compression

````python
def compress(s):
  if not s:
    return ""

  res=''
  i=0

  while i<len(s):
    count=1
    while i+1<len(s) and s[i]==s[i+1]:
      count+=1
      i+=1
    res += s[i] + str(count)
    i+=1

  return res

compress('aabbccd')
````

# 10. Longest Palindrome Substring

````python
def expand(s,left,right):
  while left>=0 and right<len(s) and s[left]==s[right]:
    left-=1
    right+=1
  return right-left-1

def lps(s):
  if not s:
    return ""
  start=end=0
  for i in range(len(s)):
    # odd
    l1 = expand(s,i,i)
    # even
    l2 = expand(s,i,i+1)
    # find max
    max_len = max(l1,l2)

    if max_len>end-start:
      start=i-(max_len-1)//2
      end = i+(max_len//2)

  return s[start:end+1]

lps('abbd')
````

# 11. Longest Common Prefix

````python
def lcp(strs):
  if not strs:
    return ""

  prefix = strs[0]

  for s in strs[1:]:
    while s[:len(prefix)]!=prefix:
      prefix = prefix[:-1]
      if prefix=='':
        return ''

  return prefix


lcp(["flower","flow","flight"])
````
