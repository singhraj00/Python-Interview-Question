```python
# Q1. Reverse only the vowels in a given string.
def rev_vowels(s):
  vowels = 'aeiouAEIOU'
  s = list(s)
  left,right=0,len(s)-1

  while left<right:
    while left<right and s[left] not in vowels:
      left+=1
    while left<right and s[right] not in vowels:
      right-=1
    # swap vowels
    s[left],s[right]=s[right],s[left]
    left+=1
    right-=1
  return ''.join(s)


print(rev_vowels('hello world'))
```

```python
# Q2. Check if two strings are anagrams without using built-in sort().
s1= 'listen'
s2 = 'silent'

def is_anagram(s1,s2):
  if len(s1)!=len(s2):
    return False

  freq = {}

  # count char in first string
  for char in s1:
    freq[char] = freq.get(char,0)+1

  #decrease count for second string
  for char in s2:
    if char not in freq:
      return False
    freq[char] -= 1
    if freq[char] <0:
      return False
  return all(value==0 for value in freq.values())

print(is_anagram(s1,s2))
print(is_anagram('pisder','spider'))
```

```python
# Q3. Find the first non-repeating character in a string.
s = 'aabbcdd'
#output = 'c'

def find_first_non_repeating(s):
  freq = {}
  for char in s:
    freq[char] = freq.get(char,0)+1

  for char in s:
    if freq[char]==1:
      print(char)
      break

  return None


find_first_non_repeating('aabbcccdef')
```

```python
# Q4. Remove all duplicate characters but keep first occurrence.
s = 'banana'
# output = 'ban'

def remove_duplicates(s):
  seen = set()
  result = []

  for char in s:
    if char not in seen:
      seen.add(char)
      result.append(char)
  return ''.join(result)


print(remove_duplicates(s))
```

```python
# Q5. Count frequency of each character without using Counter.
s = 'banana'
# output = {'b': 1, 'a': 3, 'n': 2}

def count_freq(s):
  freq = {}

  for char in s:
    freq[char]=freq.get(char,0)+1
  return freq

print(count_freq(s))
```

```python
# Q6. Rotate a list by k steps without using slicing.
lst = [1,2,3,4,5]
k = 2
# output - [4.5,1,2,3]

def rotate_list(nums,k):
  n = len(nums)
  k = k % n #normalize

  def reverse(arr,i,j):
    while i<j:
      arr[i],arr[j]=arr[j],arr[i]
      i+=1
      j-=1

  # reverse whole list
  reverse(nums,0,n-1)

  # reverse first k element
  reverse(nums,0,k-1)

  # reverse remaining elements
  reverse(nums,k,n-1)

  return nums

print(rotate_list(lst,k))
```

```python
# Q7. Find the second largest number in list (handle duplicates).

def second_largest(lst):
  largest = float('-inf')
  second_largest = float('-inf')

  for num in lst:
    if num>largest:
      second_largest = largest
      largest = num
    elif largest>num>second_largest:
      second_largest = num
  return second_largest if second_largest != float('-inf') else None


print(second_largest([10,20,20,4,5]))
```

```python
# Q8. Move all zeros to the end while keeping order of non-zeros.

def move_zeros(lst):
  pos = 0

  # Step 1: move all non-zeros forward
  for num in lst:
    if num!=0:
      lst[pos]=num
      pos+=1

  # Step 2: fill remaining places with zeros
  while pos < len(lst):
    lst[pos] = 0
    pos+=1

  return lst


print(move_zeros([10,0,20,0,30,0,40,0]))
```

```python
# Q10. Merge Two Sorted Lists (No sort(), No sorted())
a = [1,3,5]
b = [2,4,6]

# output - [1,2,3,4,5,6]

def merge_sorted(a,b):
  i=j=0
  result=[]

  while i<len(a) and j<len(b):
    if a[i]<b[j]:
      result.append(a[i])
      i+=1
    else:
      result.append(b[j])
      j+=1

  # remaining elements
  while i<len(a):
    result.append(a[i])
    i+=1

  while j<len(b):
    result.append(b[j])
    j+=1

  return result


print(merge_sorted(a,b))
```

```python
# Q11. Convert list of tuples to dictionary (handle repeated keys).

lst = [(1,2),(1,3),(2,5)]
# output: {1: [2,3], 2: [5]}

def tuple_list_to_dict(lst):
  d = {}

  for key,value in lst:
    if key not in d:
      d[key] = []
    d[key].append(value)
  return d

print(tuple_list_to_dict(lst))
```

```python
# Q12. Find the key with the maximum value in a dictionary.

data = {"a": 10, "b": 25, "c": 7}

# output: 'b'

def key_with_max_value(data):
  max_key = None
  max_value = float('-inf')

  for key,value in data.items():
    if value>max_value:
      max_value = value
      max_key = key

  return max_key

print(key_with_max_value(data))
```

```python
# Q13. Invert a Dictionary (Handle Duplicate Values)
data = {"a": 1, "b": 2, "c": 1}
# output: {1: ["a", "c"], 2: ["b"]}

def invert_dict(d):
  inverted={}

  for key,value in d.items():
    if value not in inverted:
      inverted[value]=[]
    inverted[value].append(key)
  return inverted

print(invert_dict(data))
```

```python
# Q14. Implement binary search (iterative).

def binary_search(arr,target):
  left,right=0,len(arr)-1

  while left<=right:
    mid = (left+right)//2

    if arr[mid]==target:
      return mid
    elif arr[mid]<target:
      left = mid+1
    else:
      right = mid-1
  return -1

print(binary_search([2,5,7,11,19],2))
```

```python
# binary search (recursive)

def binary_search_recursive(arr,target,left=0,right=None):
  if right is None:
    right = len(arr)-1

  # base case
  if left>right:
    return -1

  mid = (left+right)//2

  if arr[mid]==target:
    return mid
  elif arr[mid]<target:
    return binary_search_recursive(arr,target,mid+1,right)
  else:
    return binary_search_recursive(arr,target,left,mid-1)

print(binary_search_recursive([2,5,7,11],7))
```

```python
# Q15: Find the Missing Number From 1 to N
lst = [1,2,4,5]
# output - 3 missing number

# approach 1 : using math formula

def find_missing(arr):
  n = len(arr)+1
  total = n*(n+1)//2
  return total - sum(arr)


print(find_missing(lst))

# approach 2: XOR-trick

def find_missing_xor(arr):
  n = len(arr)+1
  xor1 = 0
  xor2 = 0

  for i in range(1,n+1):
    xor1 ^= i


  for num in arr:
    xor2 ^= num


  return xor1 ^ xor2

print(find_missing_xor(lst))
```

```python
# Q: Move all zeros to the end of the list (in-place, maintain order)
# Input  : [0, 1, 0, 3, 12]
# Output : [1, 3, 12, 0, 0]

def move_zeros_dsa(arr):
  j = 0 # pointer for next non-zero position

  # step1: move non-zeros elementg forward
  for i in range(len(arr)):
    if arr[i]!=0:
      arr[j]=arr[i]
      j+=1

  # step2: fill remaining elements with zeros
  while j<len(arr):
    arr[j]=0
    j+=1

  return arr


print(move_zeros_dsa([0, 1, 0, 3, 12]))
```

```python
# Q19: Maximum Subarray Sum (Kadaneâ€™s Algorithm)
# Input  : [-2,1,-3,4,-1,2,1,-5,4]
# Output : 6
# Explanation: [4, -1, 2, 1] = 6


def max_subarray(arr):
  max_sum = float('-inf')
  current_sum = 0

  for num in arr:
    current_sum += num

    if current_sum>max_sum:
      max_sum = current_sum

    if current_sum<0:
      current_sum=0

  return max_sum

print(max_subarray([-2,1,-3,4,-1,2,1,-5,4]))
```

**Kadaneâ€™s algorithm maintains a running sum of the current subarray.
If this sum becomes negative, we reset it because a negative prefix will only decrease future sum. During each step we compare and update the global maximum. This gives O(n) time and O(1) space.**

```python
# Q20. Implement stack using only list. (push/pop/top/max)

class Stack:
  def __init__(self):
    self.stack=[]
    self.max_stack=[]

  def push(self,x):
    self.stack.append(x)

    # check stack is empty or not
    if not self.max_stack or x>= self.max_stack[-1]:
      self.max_stack.append(x)

  def pop(self):
    if not self.stack:
      return None

    removed = self.stack.pop()

    # if removed element is amx , remove from max stack
    if self.max_stack and removed==self.max_stack[-1]:
      self.max_stack.pop()

    return removed

  def top(self):
    if not self.stack:
      return None
    return self.stack[-1]

  def get_max(self):
    if not self.max_stack:
      return None
    return self.max_stack[-1]

  def view_stack(self):
    return self.stack[:]

s = Stack()

s.push(3)
s.push(2)
s.push(5)
s.push(4)

print('Stack: ',s.view_stack())

print('Max: ',s.get_max())

print('removed value: ',s.pop())

print('Stack: ',s.view_stack())

print('Max: ',s.get_max())

print('removed value: ',s.pop())

print('Stack: ',s.view_stack())
```

```python
# Q21. Find pairs with given sum (optimal O(n)).

def find_pairs(nums,target):
  seen = set()
  result=[]

  for num in nums:
    complement = target-num
    if complement in seen:
      result.append((complement,num))
    seen.add(num)

  return result

print(find_pairs([2,5,7,11,4],9))
```

```python
# Q22. Reverse words in a string without using split().
# input: "hello world from python"
# output: "python from world hello"

def reverse_words(s):
  words = []
  word = ''

  for char in s:
    if char != " ":
      word += char
    else:
      if word:
        words.append(word)
      word = ""

  if word:
    words.append(word)

  # reverse manually
  left,right=0,len(words)-1
  while left<right:
    words[left],words[right]=words[right],words[left]
    left+=1
    right-=1

  # build final string
  result = ''
  for w in words:
    result += w + " "

  return result.strip()

print(reverse_words('hello world from python'))
```

```python
# Q23. Count number of vowels & consonants in a string.
def count_vowel_constants(s):
  vowels = 'aeiouAEIOU'
  vowel_count=0
  constant_count=0

  for char in s:
    if char.isalpha():
      if char in vowels:
        vowel_count += 1
      else:
        constant_count += 1
  return vowel_count, constant_count

print(count_vowel_constants('hello world'))
```

```python
# Q24. Flatten a nested list (recursive).
# Input: [1,[2,3,[4]],5] â†’ [1,2,3,4,5]

def flatten(lst):
  result=[]
  for num in lst:
    if isinstance(num,list):
      result.extend(flatten(num))
    else:
      result.append(num)
  return result

print(flatten([1,[2,3,[4]],5]))
```

```python
# generator based flatten - optimal way
def flatten_gen(lst):
  for item in lst:
    if isinstance(item,list):
      yield from flatten_gen(item)
    else:
      yield item

print(list(flatten_gen([1,[2,3],[4,[5,6]],7])))
```

```python
# Q25. Implement your own zip() function.

# Example 1
# zip([1,2,3], ['a','b','c'])
# â†’ [(1,'a'), (2,'b'), (3,'c')]

# Example 2
# zip([1,2], ['a','b','c'])
# â†’ [(1,'a'), (2,'b')]

def my_zip(*iterables):
  result=[]

  # find the len
  min_len = min(len(it) for it in iterables)

  for i in range(min_len):
    tup = tuple(it[i] for it in iterables)
    result.append(tup)

  return result

print(my_zip([1,2,3],['a','b','c']))
print(my_zip([1,2],['a','b','c']))
```

```python
# â­ Version 2 â€” True Generator Version
def my_zip_gen(*iterables):
  min_len = min(len(it) for it in iterables)

  for i in range(min_len):
    yield tuple(it[i] for it in iterables)

for x in my_zip_gen([1,2,3], ['a','b'], ['X','Y','Z']):
    print(x)
```

```python
# â­ Version 3 â€” Without using min(), without using index

def my_zip_pointer(*iterables):
  iters = [iter(it) for it in iterables]

  while True:
    try:
      yield tuple(next(it) for it in iters)
    except:
      return None

for x in my_zip_pointer([1,2,3], ['a','b'], ['X','Y','Z']):
    print(x)
```
