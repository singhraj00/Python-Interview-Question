```python
# using two pointer (in-place)
s = 'python'

s = list(s)

i,j=0,len(s)-1
while i<j:
  s[i],s[j]=s[j],s[i]
  i+=1
  j-=1

print("".join(s))
```

```python
# reverse words
sentence = 'i love banaras'

def rev_words(sentence):
  words = sentence.split()
  return  " ".join(words[::-1])

print(rev_words(sentence))
```

```python
s = 'python'

def rev_string(s):
  res = ''
  for ch in s:
    res = ch + res
  return res

print(rev_string(s))
```

```python
def rev_string_inplace(s):
  s = list(s)
  i,j=0,len(s)-1
  while i<j:
    s[i],s[j]=s[j],s[i]
    i+=1
    j-=1
  return "".join(s)

print(rev_string_inplace(s))
```

```python
def rev_words(sentence):
  words = sentence.split()
  return " ".join(words[::-1])

print(rev_words('i love python'))
```

```python
def rev_vowels(s):
  s = list(s)
  vowels = 'aeiouAEIOU'
  i,j=0,len(s)-1
  while i<j:
    if s[i] not in vowels:
      i+=1
    if s[j] not in vowels:
      j-=1
    else:
      s[i],s[j]=s[j],s[i]
      i+=1
      j-=1
  return " ".join(s)

print(rev_vowels('i love python'))
```

```python
def rev_constants(s):
  s = list(s)
  i,j=0,len(s)-1
  while i<j:
    if not s[i].isalpha():
      i+=1
    if not s[j].isalpha():
      j-=1
    else:
      s[i],s[j]=s[j],s[i]
      i+=1
      j-=1
  return " ".join(s)

print(rev_constants('i love python'))
```

```python
def rev_str_rec(s):
  if len(s)<=1:
    return s
  return rev_str_rec(s[1:]) + s[0]

print(rev_str_rec('python'))
```

```python
def rev_each_word(s):
  return " ".join(words[::-1] for words in s.split())

print(rev_each_word('i love banaras'))
```

```python
# convert string to upper case without upper()

s = 'hello'
s.upper()
```

```python
def upper(s):
  res = ''
  for char in s:
    if 'a' <= char <= 'z':
      res += chr(ord(char)-32)
    else:
      res += char
  return res

print(upper('hello'))
```

```python
s = 'HELLO'
s.lower()
```

```python
def lower(s):
  res=''
  for char in s:
    if 'A' <= char <= 'Z':
      res += chr(ord(char)+32)
    else:
      res += char
  return res

print(lower('HELLO'))
```

```python
# swap case
s = 'Hello World'
s.swapcase()
```

```python
def swap_case(s):
  res = ''
  for char in s:
    if 'a' <= char <= 'z':
      res += chr(ord(char)-32)
    elif 'A' <= char <= 'Z':
      res += chr(ord(char)+32)
    else:
      res += char
  return res

print(swap_case('Hello World'))
```

```python
# title case
s = 'hello world'
s.title()
```

```python
def title_case(s):
  words = s.split()
  return " ".join(w[0].upper() + w[1:].lower() for w in words)

print(title_case('hello world'))
```

```python
# capitalize only first letter
s = 'this is good'
s.capitalize()
```

```python
def capitalize(s):
  if not s:
    return s
  return s[0].upper() + s[1:].lower()

print(capitalize('this is good'))
```

```python
# toggle case (alternative)
s = 'hello python'
# output = 'HeLlO pYtHoN'

def alternate_swap(s):
  res = ''
  for i,char in enumerate(s):
    if i%2==0:
      res += char.lower()
    else:
      res += char.upper()
  return res

print(alternate_swap('hello python'))
```

```python
# count upper or lower case character in a string
s = 'Hello World'
# upper - 2 , lower - 8

def count_cases(s):
  lower=upper=0
  for char in s:
    if char.islower():
      lower+=1
    elif char.isupper():
      upper+=1
  return upper,lower

print(count_cases('Hello World'))
```

```python
# count vowel and constants in a string
def count_vowel_cons(s):
  vowels = 'aeiouAEIOU'
  vowels_count=cons_count=0
  for char in s:
    if char.isalpha():
      if char in vowels:
        vowels_count+=1
      else:
        cons_count+=1
  return vowels_count,cons_count


print(count_vowel_cons('Hello World'))
```

```python
s = 'Hello Python'
# capitalize first and last character each word

def cap_first_last_char(s):
  words = s.split()
  return " ".join(w[0].upper() + w[1:-1].lower() + w[-1].upper() for w in words)

print(cap_first_last_char('Hello Python'))
```

```python
# isupper() and islower() custom
s = 'hello python'

def is_upper(s):
  return all('A' <= char <= 'Z' for char in s if s.isalpha())

def is_lower(s):
  return all('a' <= char <= 'z' for char in s if s.isalpha())

print(is_upper('HELLO'))
print(is_lower('hello'))
```

```python
def custom_split(s,delimiter=' '):
  result = []
  word = ''

  for char in s:
    if char == delimiter:

      if word:
        result.append(word)
        word=''

    else:
      word += char

  if word:
    result.append(word)

  return result

print(custom_split('hello python'))
```

```python
# custom join
items = ['apple','mango','banana']
" ".join(items)
```

```python
def custom_join(lst,delimiter=' '):
  result = ' '
  for i,word in enumerate(lst):
    result += word
    if i!= len(lst) - 1:
      result += delimiter
  return result

print(custom_join(['apple','mango','banana']))
```
