```python
# using loops and join
def rev_loop(s):
  rev_chars = []
  for ch in s:
    rev_chars.insert(0,ch)
  return "".join(rev_chars)

print(rev_loop('hello'))
```

```python
# longest substring without repeating character
def longest_substring(s):
  seen = set()
  left = 0
  max_len = 0
  for right in range(len(s)):
    if s[right] in seen:
      seen.remove(s[left])
      left+=1
    seen.add(s[right])
    max_len = max(max_len,right-left+1)
  return max_len

longest_substring('abcabcbc')
```

```python
def is_anagram(a,b):
  if len(a)!=len(b):
    return False
  freq = {}
  for ch in a:
    freq[ch] = freq.get(ch,0)+1
  for ch in b:
    if ch not in freq or freq[ch]==0:
      return False
    freq[ch] -= 1
  return True


print(is_anagram('listen','silent'))
print(is_anagram('hello','world'))
```

```python
# string compression
def compress(s):
  if not s:
    return ""
  res = []
  count=1
  for i in range(1,len(s)):
    if s[i]==s[i-1]:
      count+=1
    else:
      res.append(s[i-1]+str(count))
      count=1
  res.append(s[-1]+str(count))
  return "".join(res)

print(compress('aabbccc'))
```

```python
# permute
def permute(s):
  s= list(s)
  res = []

  def backtrack(i):
    if i == len(s):
      res.append("".join(s))
      return
    for j in range(i,len(s)):
      s[i],s[j]=s[j],s[i]
      backtrack(i+1)
      s[i],s[j]=s[j],s[i]
  backtrack(0)
  return res

print(permute('abc'))
```

```python
# first unique character in a string
def unique_char(s):
  freq = {}
  for ch in s:
    freq[ch]=freq.get(ch,0)+1
  for i,ch in enumerate(s):
    if freq[ch]==1:
      return i
      break
  return -1



print(unique_char('leetcode'))
```

```python
# sort charcter by frequency
def sort_freq(s):
  freq = {}
  for ch in s:
    freq[ch] = freq.get(ch,0)+1

  # sorting
  sorted_chars = sorted(freq.items(),key=lambda x:x[1],reverse=True)

  res=''
  for ch,count in sorted_chars:
    res += ch * count

  return res


sort_freq('Aabb')
```

```python
def freq_arr_sort(arr):
  freq = {}
  for num in arr:
    freq[num]=freq.get(num,0)+1

  arr.sort(key=lambda x:(freq[x],-x))

  return arr

freq_arr_sort([1,1,2,2,2,3])
```

```python
# divide into equal parts of arr
def divide_equal(arr):
  freq = {}
  for num in arr:
    freq[num]=freq.get(num,0)+1
  for count in freq.values():
    if count % 2 != 0:
      return False

  return True


print(divide_equal([3,2,3,2,2,2]))
```

```python
# lcp

def lcp(lst):
  if not lst:
    return ''
  prefix = lst[0]
  for ch in lst[1:]:
    while ch[:len(prefix)] != prefix:
      prefix = prefix[:-1]
      if prefix == "":
        return ""
  return prefix

lcp(['flower','flight','flow'])
```

```python
# longest palibdrome substring
def lps(s):
  if not s:
    return ''
  start=end=0
  for i in range(len(s)):
    # odd
    l1 = expand(s,i,i)
    # even
    l2 = expand(s,i,i+1)
    # find max
    max_len = max(l1,l2)

    if max_len>end-start:
      start = i - (max_len-1)//2
      end = i + (max_len//2)
  return s[start:end+1]

def expand(s,left,right):
  while left >= 0 and right < len(s) and s[left]==s[right]:
    left-=1
    right+=1
  return right-left-1

lps('babad')
```
